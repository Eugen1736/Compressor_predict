В ходе анализа обучающей выборки был сделан вывод о необходимости исследования ее на выбросы и ее очистку от выбросов.

```python
df_clean = brg_df.copy()
for i in brg_dict:
  # 1. Считаем квантили и IQR
  Q1 = df_clean[i].quantile(0.25)
  Q3 = df_clean[i].quantile(0.75)
  IQR = Q3 - Q1

  # 2. Определяем границы
  lower_bound = Q1 - 3.0 * IQR
  upper_bound = Q3 + 3.0 * IQR

  # 3. Фильтруем датафрейм
  df_clean[i] = df_clean[i].clip(lower= lower_bound, upper = upper_bound)
```
